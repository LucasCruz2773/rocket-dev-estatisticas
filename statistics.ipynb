```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
    ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
    ]
}
```

```python
#Criando dataframe
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

### Média das Vendas

```python
print(f"Média das vendas: R$ {df_faturamento['valor'].mean():.2f}")
```

### Mês X Vendas

```python
#Formatando a coluna data_ref para o formato datetime
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
df_faturamento.info()
```

```python
# Formatar as datas para 'dd/mm/aaaa'
df_faturamento['data_ref_formatada'] = df_faturamento['data_ref'].dt.strftime('%d/%m/%Y')

# Exibindo como gráfico de barras
df_faturamento.sort_values(by='data_ref').plot(x='data_ref_formatada', y='valor', kind='bar')
```

```python
# Exibindo como gráfico de linha
df_faturamento.sort_values(by='data_ref').plot(x='data_ref', y='valor', kind='line')
```
